```python
import pandas as pd
import numpy as np
import matplotlib as mpl
```

```python
import matplotlib.pyplot as plt
```

```python
real_estate = pd.read_csv("real_estate_data.csv",
                          index_col = "Property")
                          
```

```python
real_estate
```

```python
property_A = real_estate[real_estate.index == "A"]
property_A
```

```python
daily_revenue = property_A["Daily_Revenue"].mean()
print("Mean of daily revenue =",daily_revenue)

monthly_cost = property_A["Monthly_Cost"].mean()
print("Mean of monthly cost =",monthly_cost)

occupancy = property_A["Occupancy_Days"].mean()
print("Mean Occupancy days =",occupancy)

calculation_A = ((daily_revenue * occupancy * 12) 
                 - (monthly_cost * 12))
print(f"Nominal yearly revenue = {calculation_A:.1f}")
```

```python
daily_revenue_std_A = property_A["Daily_Revenue"].std()
print("std of daily revenue =",daily_revenue_std_A)

monthly_cost_std_A = property_A["Monthly_Cost"].std()
print("std of monthly cost =",monthly_cost_std_A )

occupancy_std_A = property_A["Occupancy_Days"].std()
print("std Occupancy days =",occupancy_std_A)

calculation_std_A = ((daily_revenue_std_A * occupancy * 12)
                     - (monthly_cost * 12))
print(f"Standard deviation value = {calculation_std_A:.1f}")
```

```python
best_case_daily_revenue = daily_revenue + daily_revenue_std_A 
print(best_case_daily_revenue)

best_case_monthly_cost = monthly_cost + monthly_cost_std_A
print(best_case_monthly_cost)

best_case_occupancy = occupancy + occupancy_std_A
print(best_case_occupancy)

best_case_A = ((best_case_daily_revenue * best_case_occupancy * 12) 
               - (best_case_monthly_cost * 12))
print(f"The best case for Property A = {best_case_A:.1f}")
```

```python
worst_case_daily_revenue = daily_revenue - daily_revenue_std_A 
print(worst_case_daily_revenue)

worst_case_monthly_cost = monthly_cost - monthly_cost_std_A
print(worst_case_monthly_cost)

worst_case_occupancy = occupancy - occupancy_std_A
print(worst_case_occupancy)

worst_case_A = ((worst_case_daily_revenue * worst_case_occupancy * 12) 
                - (worst_case_monthly_cost * 12))
print(f"The worst case for Property A = {worst_case_A:.1f}")
```
